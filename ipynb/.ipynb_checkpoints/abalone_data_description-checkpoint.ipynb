# Abalone Data Description

```r
install.packages("dplyr")
```

```r
library("dplyr")
```

```r
abalone <- read.csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data', header=FALSE)
```

```r
colnames(abalone) = c('Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings')
```

```r
head(abalone)
```

#### Summary Statistics

The size, dimensions, and data type for Abalone.


```r
object.size(abalone)
dim(abalone)
str(abalone)
```

The summary of each feature within the dataset.

```r
summary(abalone)
```

```r
table(abalone$Sex)
barplot(table(abalone$Sex), main="Barplot of Abalone Sex", xlab="Sex", ylab="Count")
```
